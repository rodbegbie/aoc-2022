```python
# SAMPLE

INPUT = """    [D]    
[N] [C]    
[Z] [M] [P]
 1   2   3 

move 1 from 2 to 1
move 3 from 1 to 3
move 2 from 2 to 1
move 1 from 1 to 2"""
```

```python
# Part 1

stacks = [[] for x in range(9)]
stacking = True
for row in INPUT.splitlines():
    # print (row)
    if not row:
        continue
    if stacking:
        if row[1] == "1":
            # print (stacks)
            stacking = False
            continue

        pos = 0
        while row:
            box = row[1]
            if box and box != " ":
                stacks[pos].insert(0, box)
            row = row[4:]
            pos += 1
    else:
        count = int(row.split(" ")[1])
        from_stack = int(row.split(" ")[3]) - 1
        to_stack = int(row.split(" ")[5]) - 1
        # print (count, from_stack, to_stack)

        for i in range(count):
            stacks[to_stack].append(stacks[from_stack].pop())
        
        # print (stacks)


print ("".join([stack.pop() if stack else "" for stack in stacks]))

```
