```python
# Part 1

complete_overlaps = 0
for row in INPUT.splitlines():
    elf1, elf2 = row.split(",")
    (elf1_start, elf1_end) = sorted((int(x) for x in elf1.split("-")))
    (elf2_start, elf2_end) = sorted((int(x) for x in elf2.split("-")))

    if elf1_start <= elf2_start and elf1_end >= elf2_end:
        complete_overlaps += 1
        # print (row)
    elif elf2_start <= elf1_start and elf2_end >= elf1_end:
        complete_overlaps += 1
        # print (row)

print (complete_overlaps)
```

```python
# Part 2

partial_overlaps = 0
for row in INPUT.splitlines():
    elf1, elf2 = row.split(",")
    (elf1_start, elf1_end) = sorted((int(x) for x in elf1.split("-")))
    (elf2_start, elf2_end) = sorted((int(x) for x in elf2.split("-")))

    if elf1_start <= elf2_end and elf1_end >= elf2_start:
        partial_overlaps += 1
        # print (row)
    elif elf2_start <= elf1_end and elf2_end >= elf1_start:
        partial_overlaps += 1
        # print (row)

print (partial_overlaps)
```
