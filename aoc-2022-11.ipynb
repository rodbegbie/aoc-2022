```python
# SAMPLE

INPUT = """Monkey 0:
  Starting items: 79, 98
  Operation: new = old * 19
  Test: divisible by 23
    If true: throw to monkey 2
    If false: throw to monkey 3

Monkey 1:
  Starting items: 54, 65, 75, 74
  Operation: new = old + 6
  Test: divisible by 19
    If true: throw to monkey 2
    If false: throw to monkey 0

Monkey 2:
  Starting items: 79, 60, 97
  Operation: new = old * old
  Test: divisible by 13
    If true: throw to monkey 1
    If false: throw to monkey 3

Monkey 3:
  Starting items: 74
  Operation: new = old + 3
  Test: divisible by 17
    If true: throw to monkey 0
    If false: throw to monkey 1"""
```

```python
# Part 1

monkeys = []
common_factor = 1

class Monkey:
    monkey_id = None
    operator = None
    operation_number = None
    items = None
    test_divisible = None
    true_monkey = None
    false_monkey = None

    inspect_count = 0

    def catch_item(self, item):
        self.items.append(item)

    def inspect(self, divide=True):
        # print ("Monkey", self.monkey_id)
        for item in self.items:
            self.inspect_count += 1
            # print ("  Inspecting", item)
            num = item if self.operation_number == -1 else self.operation_number
            if self.operator == "+":
                item += num
            else:
                item *= num
            # print ("    Worry up to", item)
            if divide:
                item //= 3
            else:
                item %= common_factor
            # print ("    Worry down to", item)

            if item % self.test_divisible:
                # print ("    Not divisible. Throw to", self.false_monkey)
                monkeys[self.false_monkey].catch_item(item)
            else:
                # print ("    Divisible. Throw to", self.true_monkey)
                monkeys[self.true_monkey].catch_item(item)
        
        self.items = []

def get_monkeys_from_input():
    global monkeys, common_factor

    monkeys = []
    common_factor = 1
    rows = INPUT.splitlines()
    while rows:
        monkey_rows = rows[:6]
        rows = rows[6:]
        if rows:
            rows = rows[1:]
        
        monkey = Monkey()
        monkeys.append(monkey)

        monkey.monkey_id = int(monkey_rows[0][7])
        monkey.items = [int(x) for x in monkey_rows[1].split(": ")[1].split(", ")]
        monkey.operator = monkey_rows[2].split(" ")[6]
        monkey.operation_number = -1 if monkey_rows[2].split(" ")[7] == "old" else int(monkey_rows[2].split(" ")[7])
        monkey.test_divisible = int(monkey_rows[3].split(" ")[5])
        common_factor *= monkey.test_divisible
        monkey.true_monkey = int(monkey_rows[4].split(" ")[9])
        monkey.false_monkey = int(monkey_rows[5].split(" ")[9])

get_monkeys_from_input()

for _ in range(20):
    for monkey in monkeys:
        monkey.inspect()

counts = [monkey.inspect_count for monkey in monkeys]
counts = sorted(counts, reverse=True)
print (counts[0] * counts[1])


```
