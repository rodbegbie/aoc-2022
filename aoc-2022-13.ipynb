```python
# Part 1
import json

RIGHT = 1
WRONG = -1

index = 0
index_sum = 0

def is_right_order(left, right):
    if type(left) == int and type(right) == int:
        if left < right:
            # print ("INT_RIGHT")
            return RIGHT
        elif left > right:
            # print ("INT_WRONG")
            return WRONG

    elif type(left) == list and type(right) == list:
        for i in range(max((len(left), len(right)))):
            if i >= len(left):
                # print ("LIST_RIGHT")
                return RIGHT
            elif i >= len(right):
                # print ("LIST_WRONG")
                return WRONG
            else:
                check_element = is_right_order(left[i], right[i])
                if check_element in [RIGHT, WRONG]:
                    return check_element
    
    elif type(left) == list and type(right) == int:
        return is_right_order(left, [right])
    elif type(left) == int and type(right) == list:
        return is_right_order([left], right)

    


rows = INPUT.splitlines()
while rows:
    index += 1
    (pair1, pair2) = rows[:2]
    rows = rows[2:]
    if rows:
        #skip blank
        rows = rows[1:]
    
    if is_right_order(json.loads(pair1), json.loads(pair2)) == RIGHT:
        index_sum += index
        print ("MATCH", index, index_sum)

print (index_sum)
```
